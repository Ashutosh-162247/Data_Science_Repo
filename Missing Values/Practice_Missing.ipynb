```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('synthetic_dataset_with_missing_values.csv')
df.head()
```

```python
## find the missing values
df.isnull()
```

```python
df.isnull().sum()
```

```python
## there are 20 values in each column having the null values
##let's start with Age
##find the mean of age
df['Age'].mean()
```

```python
mean=(df['Age'].sum())/(df['Age'].count())
print(mean)
```

```python
## fill the missing values with mean
df['Age'].fillna(df['Age'].mean, inplace=True)
```

```python
##now all the null valued are filled
df.isnull().sum()
```

```python
df.head()
```

```python
df['Salary'].mean()
```

```python
df['Salary'].fillna(df['Salary'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
## for Department this is categorial value
df.groupby(['Department']).count()
```

```python
df['Department'].fillna(df['Department'].mode()[0],inplace=True)
```

```python
df.groupby(['Department']).count()
```

```python
df.isnull().sum()
```

```python
df['Experience'].fillna(df['Experience'].mean, inplace=True)
df['Education'].fillna(df['Education'].mean, inplace=True)
```

```python
df.isnull().sum()
```
