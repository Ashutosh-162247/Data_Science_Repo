```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('linear_regression_40rows.csv')
```

```python
df.head()
```

```python
## Analze the data 
sns.pairplot(df)
```

```python
df.corr()
```

```python
## Here Size_sqrt and Rooms are input features and Price_k is output feature.
##so sepate the input and output feature
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
y
```

```python
## Test Train Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=7)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression(n_jobs=-1)
```

```python
regression.fit(X_train,y_train)
```

```python
regression.fit(X_train,y_train)
```

```python
y_pred=regression.predict(X_test)
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error,mean_squared_error
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
rmse=np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
```

```python
X_train.size,y_train.size,X_test.size,y_test.size
```

```python
X.size,y.size
```

```python

plt.figure(figsize=(6,5))
plt.scatter(y_test, y_pred, c='navy', alpha=0.8, label='Predictions')
minv, maxv = min(y_test.min(), y_pred.min()), max(y_test.max(), y_pred.max())
plt.plot([minv, maxv], [minv, maxv], 'r--', label='Ideal (y=x)')
plt.xlabel("True Price_k")
plt.ylabel("Predicted Price_k")
plt.title("Linear Regression: True vs Predicted")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
